# Task 04: Basic Data Visualization

```python
# Histogram of salaries
plt.hist(df['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart of department proportions
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(10)
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%')
plt.title('Proportion of Employees in Top 10 Departments')
plt.show()
```

# Task 05: Grouped Analysis

```python
grouped_data = df.groupby('JobTitle')['BasePay'].describe()
print("\nSummary statistics by job title:")
print(grouped_data.head())
```

# Task 06: Simple Correlation Analysis

```python
plt.scatter(df['BasePay'], df['TotalPay'], alpha=0.5)
plt.title('Scatter Plot of BasePay vs TotalPay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```

```python
correlation_coefficient = df['BasePay'].corr(df['TotalPay'])
print("Correlation coefficient between BasePay and TotalPay:", correlation_coefficient)
```

# Task 07: Summary of Insights

3. In the Scatter Plot of BasePay versus TotalPay, we notice that when BasePay goes up, TotalPay also goes up. This is because BasePay is a big part of TotalPay, and when it's higher, employees overall get paid more. Also, when BasePay is higher, employees might get more extra pay or benefits, making their total pay even higher.
